### Do it! 실습 | 산점도 만들기

```python
import pandas as pd
mpg = pd.read_csv("/Users/heohyeonjun/Desktop/We_IT_5th/weit/Data/mpg.csv") # 데이터 불러오기
```

```python
# x축은 displ, y축은 hwy를 나타낸 산점도 만들기
import seaborn as sns
sns.scatterplot(x="displ", y="hwy", data=mpg)
```

```python
# x축 범위 3~6으로 제한
sns.scatterplot(data=mpg, x="displ", y='hwy').set(xlim = (3,6), ylim = (10, 30))
```

```python
#drv별로 표식 색깔 다르게 표현
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy', hue='drv')
```
