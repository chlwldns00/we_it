### Do it! 실습 | 여러 조건을 충족하는 행 추출하기

### Do it! 실습 | 문자 변수를 이용해 조건에 맞는 행 추출하기

### mpg 데이터를 이용해 분석 문제를 해결해 보세요.

#### Q1. 자동차 배기량에 따라 고속도로 연비가 다른지 알아보려고 한다. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비) 평균이 더 높은지 알아보자.

```python
# displ(배기량)이 4이하인 자동차와 5이상인 자동차를 추출한다.
displ_under_4 = mpg.query("displ <= 4")
displ_under_5 = mpg.query("displ >= 5")

# 평균을 구하자.
under_4_mean = displ_under_4['hwy'].mean()
under_5_mean = displ_under_5['hwy'].mean()
```

#### Q2. 자동차 제조 회사에 따라 도시 연비가 어떻게 다른지 알아보려고 한다. 'audi와 'toyota'중 어느 manufacturer(자동차 제조 회사)의 cty(도시 연비) 평균이 더 높은지 알아보자.

```python
# audi와 toyota 추출
audi = mpg.query('manufacturer == "audi"')
toyota = mpg.query('manufacturer == "toyota"')
```

```python
# 도시 연비 구하기
audi['cty'].mean()
```

```python
# 도시 연비 구하기
toyota['cty'].mean()
```
